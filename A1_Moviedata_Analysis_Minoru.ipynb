```python
ratings_mapping = {
   3: '1_very_poor',     # 3.0-3.9
   4: '2_poor',          # 4.0-4.9
   5: '3_fair',          # 5.0-5.9
   6: '4_good',          # 6.0-6.9
   7: '5_very_good',     # 7.0-7.9
   8: '6_excellent'      # 8.0-8.9
}

# Map to categories by rounding down to integer
movie['cat_rating'] = movie['Ratings'].apply(lambda x: ratings_mapping[int(x)])
```

```python
# prompt: create bar chart of ca_rating from 1_poor to 6_excellent

plt.figure(figsize=(10, 6))
sns.countplot(x='cat_rating', data=movie, order=['1_very_poor', '2_poor', '3_fair', '4_good', '5_very_good', '6_excellent'])
plt.title('Distribution of Movie Ratings (Categorical)')
plt.xlabel('Ratings Categories')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# prompt: 20142015cat_rating.

# Convert 'Release Date' to datetime objects
movie['Release Date'] = pd.to_datetime(movie['Release Date'])

# Extract the year from the 'Release Date'
movie['Year'] = movie['Release Date'].dt.year

# Group by year and category rating, then count the number of movies
year_cat_rating_counts = movie.groupby(['Year', 'cat_rating'])['Title'].count().unstack()

# Print the result
print(year_cat_rating_counts)

# Visualize the comparison
year_cat_rating_counts.plot(kind='bar', figsize=(12, 6))
plt.title('Number of Movies by Year and Rating Category')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.xticks(rotation=0)
plt.legend(title='Rating Category')
plt.show()
```

### Data Viz

```python
# prompt: Gross

# Correlation analysis between Gross and multiple variables
correlation_matrix = movie[['Gross', 'Budget', 'Ratings', 'Year']].corr()

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Reds', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Scatter plot to visualize the relationship between Gross and Budget
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Budget', y='Gross', data=movie)
plt.title('Gross vs. Budget')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()

# Scatter plot to visualize the relationship between Gross and Ratings
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Ratings', y='Gross', data=movie)
plt.title('Gross vs. Ratings')
plt.xlabel('Ratings')
plt.ylabel('Gross')
plt.show()

# Scatter plot to visualize the relationship between Gross and Year
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Year', y='Gross', data=movie)
plt.title('Gross vs. Year')
plt.xlabel('Year')
plt.ylabel('Gross')
plt.show()

# You can further analyze the relationship using regression analysis or other statistical methods.
# For example, you can fit a linear regression model to see how Gross is affected by Budget, Ratings, and Year.

# Example of linear regression using statsmodels
import statsmodels.formula.api as sm

# Fit a linear regression model
model = sm.ols('Gross ~ Budget + Ratings + Year', data=movie).fit()

# Print the model summary
print(model.summary())
```

year 2014 and 2015 mix

Findings from the data analysis:<br>

1. Correlation Matrix Interpretation:

Gross vs Budget: 0.70 (strong positive correlation)<br>
Gross vs Ratings: 0.39 (weak to moderate positive correlation)<br>
Gross vs Year: 0.12 (negligible correlation)<br>

<br>
2. Scatter Plot Analysis:<br>

Budget vs Gross: Clear upward trend indicating higher budgets tend to generate higher revenues<br>
Ratings vs Gross: Weak positive correlation; higher ratings don't necessarily guarantee higher revenue<br>
Year vs Gross: No clear trend based on release year<br>

<br>
3. Regression Analysis (OLS Results):<br>


R-squared: 0.528 (model explains about 53% of variance)<br>
Budget: p-value 0.000 (statistically significant)<br>
Ratings: p-value 0.000 (statistically significant)<br>
Year: p-value 0.131 (not statistically significant)<br>

<br>
Key Findings:<br>

Budget is the strongest predictor of revenue<br>
Ratings have moderate influence<br>
Release year has minimal impact<br>
The model explains approximately 53% of revenue variance<br>

```python
# Split data for 2014 and 2015
movie_2014 = movie[movie['Year'] == 2014]
movie_2015 = movie[movie['Year'] == 2015]

# Analyze each year separately
for year, data in [('2014', movie_2014), ('2015', movie_2015)]:
   print(f"\nAnalysis for {year}")

   # Create correlation matrix
   correlation_matrix = data[['Gross', 'Budget', 'Ratings']].corr()

   # Visualize correlation matrix
   plt.figure(figsize=(10, 8))
   sns.heatmap(correlation_matrix, annot=True, cmap='Reds', fmt=".2f")
   plt.title(f'Correlation Matrix - {year}')
   plt.show()

   # Scatter plot for Budget vs Gross
   plt.figure(figsize=(8, 6))
   sns.scatterplot(x='Budget', y='Gross', data=data)
   plt.title(f'Gross vs. Budget - {year}')
   plt.xlabel('Budget')
   plt.ylabel('Gross')
   plt.show()

   # Scatter plot for Ratings vs Gross
   plt.figure(figsize=(8, 6))
   sns.scatterplot(x='Ratings', y='Gross', data=data)
   plt.title(f'Gross vs. Ratings - {year}')
   plt.xlabel('Ratings')
   plt.ylabel('Gross')
   plt.show()

   # Perform regression analysis
   model = sm.ols('Gross ~ Budget + Ratings', data=data).fit()
   print("\nRegression Results:")
   print(model.summary().tables[1])
```
